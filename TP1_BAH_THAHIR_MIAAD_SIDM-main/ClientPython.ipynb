```python
import tkinter as tk
import socket
import threading

def recevoir_messages():
    while True:
        try:
            message = client_socket.recv(1024).decode()
            liste_messages.insert(tk.END, message)
        except ConnectionAbortedError:
            break

def envoyer_message():
    message = champ_saisie.get()
    client_socket.sendall(message.encode())
    champ_saisie.delete(0, tk.END)

def connecter():
    global client_socket
    host = champ_host.get()
    port = int(champ_port.get())
    
    try:
        client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        client_socket.connect((host, port))
        threading.Thread(target=recevoir_messages).start()
        bouton_connecter.config(state=tk.DISABLED)
        champ_host.config(state=tk.DISABLED)
        champ_port.config(state=tk.DISABLED)
        champ_saisie.config(state=tk.NORMAL)
        bouton_envoyer.config(state=tk.NORMAL)
    except ConnectionRefusedError:
        liste_messages.insert(tk.END, "Erreur de connexion au serveur")

fenetre = tk.Tk()
fenetre.title("Client Chat")
fenetre.geometry("800x600")  
fenetre.resizable(True, True) 
# Cadre supérieur
cadre_haut = tk.Frame(fenetre, bg="orange")
cadre_haut.pack(pady=20)

label_host = tk.Label(cadre_haut, text="Host:")
label_host.pack(side=tk.LEFT)

champ_host = tk.Entry(cadre_haut, width=90)
champ_host.pack(side=tk.LEFT)

label_port = tk.Label(cadre_haut, text="Port:")
label_port.pack(side=tk.LEFT)

champ_port = tk.Entry(cadre_haut,)
champ_port.pack(side=tk.LEFT)

bouton_connecter = tk.Button(cadre_haut, text="Connecter", command=connecter)
bouton_connecter.pack(side=tk.LEFT)

# Liste des messages
cadre_Centre = tk.Frame(fenetre)
cadre_Centre.pack(pady=20)
liste_messages = tk.Listbox(cadre_Centre, width=100)
liste_messages.pack(pady=20)

# Cadre inférieur
cadre_bas = tk.Frame(fenetre)
cadre_bas.pack(pady=10)

label_Message = tk.Label(cadre_bas, text="Message:")
label_Message.pack(side=tk.LEFT)

champ_saisie = tk.Entry(cadre_bas, width=90, state=tk.DISABLED)
champ_saisie.pack(side=tk.LEFT)

bouton_envoyer = tk.Button(cadre_bas, text="Envoyer", state=tk.DISABLED, command=envoyer_message)
bouton_envoyer.pack(side=tk.LEFT)

fenetre.mainloop()
```
